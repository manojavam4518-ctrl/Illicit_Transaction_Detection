```python
import pandas as pd
import os
import zipfile

# --- FINAL DATA LOADING SOLUTION (Corrected Path) ---

DATASET_SLUG = "ellipticco/elliptic-data-set"
DOWNLOAD_NAME = "elliptic-data-set.zip"
UNZIP_DIR = "elliptic_data_files"
# NEW SUBDIRECTORY FOUND AFTER UNZIPPING
SUB_DIR = "elliptic_bitcoin_dataset"

# 1. Download and Unzip steps (assuming they ran successfully before this)
# ... (No need to run the !kaggle download and zipfile.ZipFile(...) lines again if the files are there)

# 2. Define the corrected paths
# FIX: Include the SUB_DIR in the path construction
FEATURES_FILE = os.path.join(UNZIP_DIR, SUB_DIR, "elliptic_txs_features.csv")
CLASSES_FILE = os.path.join(UNZIP_DIR, SUB_DIR, "elliptic_txs_classes.csv")

# 3. Load the DataFrames
try:
    print("\nAttempting to load data with corrected Pandas paths...")

    # Load the features file (header=None is crucial)
    features_df = pd.read_csv(FEATURES_FILE, header=None)
    features_df = features_df.rename(columns={0: 'txId', 1: 'Time step'})

    # Load the class labels file
    classes_df = pd.read_csv(CLASSES_FILE)

    print("\nData Loading Complete! ğŸ‰")
    print(f"Features Data Shape: {features_df.shape}")
    print(f"Classes Data Shape: {classes_df.shape}")
    print("\nFeatures Data Head:")
    print(features_df.head())

except FileNotFoundError:
    print("\nCRITICAL FAILURE: Files still not found. Check the file names and the SUB_DIR variable.")
```

```python
## 3.1 Merge Features and Classes
# Merge based on the common transaction ID ('txId')
df = pd.merge(features_df, classes_df, on='txId', how='left')

TARGET_COLUMN = 'class'

# --------------------------------------------------------------------------------------
# CRITICAL FIX: Filter out ALL unlabeled/unknown transactions explicitly
# The original dataset labels are '1' (Illicit), '2' (Licit), and 'unknown' (string).
# We must ensure 'unknown' is removed before conversion.

# Filter out rows where the class is a specific string 'unknown' OR is NaN
labeled_df = df[df[TARGET_COLUMN].isin(['1', '2'])].copy()
# Note: .copy() is added to avoid a SettingWithCopyWarning later
# --------------------------------------------------------------------------------------

# Select Features (X) and Target (y)
feature_columns = labeled_df.columns.drop(['txId', 'class'])
X = labeled_df[feature_columns]
y = labeled_df[TARGET_COLUMN]

# Map classes: '2' (Licit) -> 0, '1' (Illicit) -> 1
y = y.replace({'1': 1, '2': 0})

# ------------------------------------------------------------------
# CRITICAL FIX 1: Convert the target series 'y' to a numerical type (integer)
# This is now SAFE because 'unknown' has been filtered out.
y = y.astype(int)
# ------------------------------------------------------------------


print("Target Variable Distribution (0=Licit, 1=Illicit):")
print(y.value_counts())

## 3.2 Scaling and Train-Test Split
# CRITICAL FIX 2 (from previous error): Convert all feature names (columns) to strings
X.columns = X.columns.astype(str)
# ------------------------------------------------------------------


# Scale the features (StandardScaler is vital for DNNs)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data (stratified to maintain the 0/1 ratio in both sets)
# This should now work as y is pure integer type.
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f"\nTraining set shape: {X_train.shape}")
print(f"Test set shape: {X_test.shape}")
```

```python
## 4.1 Define the Deep Neural Network Model
# Ensure you have imported from tensorflow.keras.models/layers at the start of your notebook!
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

input_dim = X_train.shape[1] # Should be 167 (Time step + 166 features)
hidden_units_1 = 512
hidden_units_2 = 256
hidden_units_3 = 128

model = Sequential([
    # Input Layer + First Hidden Layer
    Dense(hidden_units_1, activation='relu', input_shape=(input_dim,)),
    Dropout(0.3),

    # Second Hidden Layer
    Dense(hidden_units_2, activation='relu'),
    Dropout(0.3),

    # Third Hidden Layer
    Dense(hidden_units_3, activation='relu'),

    # Output Layer (1 neuron with sigmoid for binary classification)
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

print("\n--- Model Architecture and Parameter Count ---")
model.summary()

## 4.2 Train the Model
print("\nStarting Deep Learning model training...")
history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=128,
    validation_split=0.1,
    verbose=1
)
print("Model training finished!")

## 4.3 Evaluation
# Predict probabilities and convert to class predictions (0 or 1)
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

# Evaluation Metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

print("\n--- Model Evaluation (Illicit Transaction Detection) ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, zero_division=0))

# Visualize the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues',
            xticklabels=['Licit (0)', 'Illicit (1)'], yticklabels=['Licit (0)', 'Illicit (1)'])
plt.title('Confusion Matrix: Illicit Transaction Detection')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.show()
```
